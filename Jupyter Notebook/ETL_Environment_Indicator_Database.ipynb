## Confirm Country Code


Steps: 

--Iterate through dataframes
1. Need to iterate through dataframes, thus used a dictionary to maintain the 'type' of the dataframe while putting in an structured format by having the keys in numerical order. 
2. Created a text list of the dataframes in the dict so I could print the name of the data, but not the dataframe itself

-- count the unique country codes
1. If each country has more than 1 unqiue country code then the boolean will print false. 
Thus meaning that df countries have to be reviewed and edited to become unique



```python
dict_of_df = {0 : endangered_species_data, 1 : water_quality_data, 2 : air_quality_data, 3: wdi, 4 : health_risk_data, 5 : waste_treatment_data}
list_of_df = ['endangered_species_data','water_quality_data', 'air_quality_data', 'wdi','health_risk_data','waste_treatment_data']
```

```python
## count unique country code per count.  if value > 1 then print series row otherwise dropna

i = 0 
while i <= 5:
    for i in dict_of_df: ## i is acting as key value
        df_name = list_of_df[i]
        country_code = dict_of_df[i].groupby('Country')['COU'].nunique()  ##print 'value' of dictionary using key value
        check = country_code.where(country_code > 1).dropna().empty
        print(f"Dataframe: {df_name} Country Code Check (True/Good) (False/Error) = {check}")
        i = i + 1
```

#### Prepare Datasets to  be mergered

1. Print only unique rows of COU per country
2. Unique function prints values into a list, thus we need to make them into strings so they can be located when we merge the datasets together

```python
## Create new df of country codes 

i = 0 
code_list = []
while i <= 5:
    for i in dict_of_df: ## i is acting as key value
        df_name = list_of_df[i] ## records which database we are iterating through 
        cou = dict_of_df[i][['Country','COU']].groupby('Country')['COU'].unique().reset_index(name="COU").sort_values(by='Country')
        """
        -iterate through dataframes
        -selects two columns for country code
        -print only unique list of COU, 
        -set series values name to 'COU' to make dataframe
        - sort countries name in alphabetical order
        
        """        
        code_list.append(cou) ## append dataframe to list
        i= i+1 ## loop through next dataframe 
        print(f"Completed Concat of {df_name}") #state which dataframes are complete
    code_df = pd.concat(code_list,ignore_index=True) # combine as 1 long list of unique countries
```

```python
code_df.tail()
```

```python
code_df['COU'][0][0]
```

```python
## to extract COU code from the indivudal list per row
### create items in row as a list and reassign it to column 'COU'

i=0
country_code = []

while i <= (len(code_df['COU'])-1): ##sets limit per dataframe
    for x in code_df['COU']:
        country_code.append(code_df['COU'][i][0])
        i=i+1
```

```python
## reassign COU to new country code
code_df['COU'] = country_code
```

```python
unique_code_df = code_df.groupby('Country')['COU'].unique().reset_index(name="COU").sort_values(by='Country')
```

```python
## to extract COU code from the indivudal list per row
### create items in row as a list and reassign it to column 'COU'

i=0
country_code_2nd = []

while i <= (len(unique_code_df['COU'])-1): ##sets limit per dataframe
    for x in unique_code_df['COU']:
        country_code_2nd.append(unique_code_df['COU'][i][0])
        i=i+1
```

```python
unique_code_df['COU'] = country_code_2nd
```

```python
unique_code_df.head()
```

```python
check_codes = unique_code_df.groupby('Country')['COU'].nunique()
codes = check_codes.where(check_codes > 1).dropna()
final_check = list(codes.index)
final_check
```

### <font color= blue> Unsafe Water Source </font>

UOM for health risk is DALYs. 

DALY stands for Disability-Adjusted Life Year (DALY). 

Per sample size of population they calculate the cause of death and the amount of years that person was suppose to live depending on thier age. 
The cause of death can be premature due to disease or disability.
Disability factors are applied a weighted average according to the severity and length of disability.

This metric was developed and is calculate by the Global Burden of Disease Study and World Health Organization. 


source of uom metric is: https://www.who.int/healthinfo/global_burden_disease/metrics_daly/en/

We'll review country's DALYs Avgerage over 17 yrs per Risk Factor.
